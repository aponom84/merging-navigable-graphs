```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results/results.csv')
```

```python
for alg, marker in zip(df.algorithm.unique() , '>o+*^'):
    # plt.scatter(df[df.algorithm == alg]['search distance count'], df[df.algorithm == alg]['5@recall'], label=alg)
    plt.plot(df[df.algorithm == alg]['search distance count'],
             df[df.algorithm == alg]['5@recall'], 
             label=alg,
             marker=marker,
             markersize=8)
plt.xlabel("Distance count")
plt.ylabel("5@recall")
plt.legend()
```

```python
_df = df[df.algorithm!='merge naive'][['algorithm', 'merge distance count']].drop_duplicates()
fig, ax = plt.subplots()
bar_colors = ['tab:green', 'tab:blue', 'tab:red', 'tab:orange']
ax.bar(_df.algorithm, _df['merge distance count'], label=_df.algorithm, color=bar_colors)
ax.set_ylabel('Distance count')
ax.set_title('Merging algorithms efforts')
ax.legend(title='Merging algorithms')

plt.show()
```
