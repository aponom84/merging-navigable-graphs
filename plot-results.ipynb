```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Merge Distance Count

```python
df = pd.read_csv('results/results_16m32m0.csv').head(3)
```

```python
_fig, ax = plt.subplots()
bar_colors = ['tab:green', 'tab:blue', 'tab:red', 'tab:orange']
# ax.bar(_df.algorithm, _df['merge distance count'], label=_df.algorithm, color=bar_colors)
ax.bar(df.algorithm, df['merge distance count'], label=df.algorithm+' '+df.params, color=bar_colors)
ax.set_ylabel('Distance count')
ax.set_title('Merging algorithms efforts')
ax.legend(title='Merging algorithms')

plt.show()
```

# 5@recall/cistance-count 

```python
recall_cols = [c for c in df if c[-6:]=='recall']
dist_count_cols = [c for c in df if c[-10:]=='dist count']
```

```python
for (i, row), marker in zip(df.iterrows(), '>o+*^'):
    plt.plot(row[dist_count_cols],
             row[recall_cols],
             label=row.algorithm,
             marker=marker,
             markersize=8)
plt.xlabel("Distance count")
plt.ylabel("5@recall")
plt.legend()
```
